# 🚀 Projeto 1 - Sistema Comercial Completo
Objetivo: Uma empresa deseja gerar um relatório completo de vendas.
1. Mostre todas as vendas.
2. Calcule
    * valor total vendido
    * quantidade total de itens vendidos
    * ticket médio por venda
    * maior venda
    * menor venda
3. Conte quantas vendas cada vendedor realizou
4. Calcule quanto cada vendedor faturou.
5. Calcule a média das vendas.
6. Classifique os vendedores:
    * Mais de R$10000 = Excelente
    * Mais de R$7000 = Bom
    * Caso contrário = Regular
7. Descubra:
    * vendedor que mais vendeu
    * vendedor que menos vendeu
8. Agrupe por cidade.
9. Monte um resumo final:
    * TOTAL VENDIDO
    * MAIOR VENDA
    * MENOR VENDA
    * MÉDIA
    * ITENS VENDIDOS
    * VENDEDOR DESTAQUE
    * CIDADE DESTAQUE

```python
vendas = [
    {"Vendedor":"Ana","Cidade":"Curitiba","Produto":"Notebook","Quantidade":2,"Valor":3500},
    {"Vendedor":"Carlos","Cidade":"Joinville","Produto":"Mouse","Quantidade":10,"Valor":80},
    {"Vendedor":"Maria","Cidade":"Curitiba","Produto":"Monitor","Quantidade":3,"Valor":1200},
    {"Vendedor":"Ana","Cidade":"Curitiba","Produto":"Teclado","Quantidade":5,"Valor":150},
    {"Vendedor":"Carlos","Cidade":"Joinville","Produto":"Notebook","Quantidade":1,"Valor":3500},
    {"Vendedor":"João","Cidade":"Blumenau","Produto":"Mouse","Quantidade":15,"Valor":80},
    {"Vendedor":"Maria","Cidade":"Curitiba","Produto":"Notebook","Quantidade":2,"Valor":3500},
    {"Vendedor":"Ana","Cidade":"Curitiba","Produto":"Monitor","Quantidade":1,"Valor":1200}
]


venda_total = 0
quantidade_total = 0
ticket_médio = 0
maior_venda = vendas[0]["Valor"]
menor_venda = vendas[0]["Valor"]
contagem = {}

# print("=" * 40)
# print("RELATÓRIO VENDAS")
# print("=" * 40)

for venda in vendas:
  nome = venda["Vendedor"]
  cidade = venda["Cidade"]
  produto = venda["Produto"]
  quantidade = venda["Quantidade"]
  venda = venda["Valor"]
  venda_vendedor = venda * quantidade
  venda_total += venda_vendedor
  quantidade_total += quantidade

  if venda_vendedor > maior_venda:
    maior_venda = venda_vendedor

  if venda_vendedor < menor_venda:
    menor_venda = venda_vendedor

  if nome not in contagem:
    contagem[nome] = 1
  else:
    contagem[nome] += 1

  # print(f"\nVendedor(a): {nome}")
  # print(f"Cidade: {cidade}")
  # print(f"Produto: {produto}")
  # print(f"Quantidade: {quantidade}")
  # print(f"Valor: R$ {venda:.2f}")
  # print(f"Venda Total: R$ {venda_vendedor:.2f}")

ticket_médio = venda_total / quantidade_total
# print()
# print("=" * 40)
print("RESUMO VENDAS")
print(f"Venda Total ..... R$ {venda_total:.2f}")
print(f"Quantidade Total ..... {quantidade_total}")
print(f"Ticket Médio ..... R$ {ticket_médio:.2f}")
print(f"Maior Venda ..... R$ {maior_venda:.2f}")
print(f"Menor Venda ..... R$ {menor_venda:.2f}")
print(f"\nQuantidade de Vendas por vendedor:")
for nome,qtd in contagem.items():
  print(f"- {nome} | {qtd} Vendas")

  
```
