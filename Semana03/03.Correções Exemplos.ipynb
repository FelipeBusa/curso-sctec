## Exercícios da Semana
* Fácil: Crie uma lista com 5 itens e imprima o segundo e o último usando índices.
* Médio: Crie uma lista de dicionários representando produtos e percorra com for imprimindo nome e preço de cada um.
* Avançado: Crie um carrinho de compras (lista de dicionários com produto/preço/quantidade) e calcule o valor total.
