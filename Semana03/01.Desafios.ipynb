Exercício 1 — Vendas por Produto (Nível Médio)

Uma loja registrou as seguintes vendas:

produtos = ["Notebook", "Mouse", "Teclado", "Monitor"]
quantidades = [8, 25, 12, 5]

Percorra as listas e mostre:

Notebook -> 8 unidades
Mouse -> 25 unidades
...

Ao final informe:

Quantos produtos venderam mais de 10 unidades.
Exercício 2 — Cadastro de Funcionário (Nível Médio)

Crie um dicionário para armazenar:

Nome
Cargo
Salário
Tempo de empresa

Exemplo:

funcionario = {
    "nome": "Carlos",
    "cargo": "Analista",
    "salario": 4800,
    "anos_empresa": 4
}

Mostre todas as informações utilizando f-strings.

Depois faça:

Se salário for maior que 5000
Salário acima da média.
Caso contrário
Salário dentro da média.
Exercício 3 — Controle de Estoque (Nível Médio)

Uma empresa possui:

estoque = {
    "Notebook": 18,
    "Mouse": 4,
    "Monitor": 7,
    "Teclado": 25
}

Percorra o dicionário.

Para cada produto:

Se quantidade menor que 5

Reposição urgente

Caso contrário

Estoque OK
Exercício 4 — Análise Comercial (Nível Difícil)

As vendas do mês foram:

vendas = [1250, 980, 2300, 450, 1750, 890, 3200]

Percorra a lista e informe:

Quantas vendas foram maiores que R$1000
Quantas vendas foram menores que R$1000
Quantas vendas são valores pares (utilize %)

Ao final mostre o total de cada categoria.

Exercício 5 — Ranking de Lojas (Nível Difícil)

lojas = [

]

Percorra a lista.

Meta atingida

Caso contrário

Meta não atingida
Exercício 6 — Auditoria de Clientes (Nível Difícil)

Você recebeu:

clientes = [
    ("Felipe", 5),
    ("Ana", 1),
    ("João", 0),
    ("Carlos", 3),
    ("Maria", 8)
]

Cada tupla representa:

(nome, quantidade de compras)

Mostre:

Felipe realizou 5 compras.

Depois classifique:

0 compras

Cliente Inativo

1 até 3

Cliente Ativo

Acima de 3

Cliente VIP

Exercício 8 — Controle de Chamados de TI (Desafiador)

Uma empresa possui:

chamados = {
    "Aberto": 14,
    "Em andamento": 7,
    "Resolvido": 31,
    "Cancelado": 2
}

Percorra o dicionário.

Para cada situação:

Se quantidade maior que 10

Alta demanda

Caso contrário

Fluxo normal

Depois informe quantos chamados ainda não foram resolvidos.

(Considere "Aberto" e "Em andamento" como não resolvidos.)

Exercício 9 — Controle Financeiro (Desafiador)

Uma empresa registrou as movimentações:

movimentacoes = [
    ("Receita", 2500),
    ("Despesa", 1200),
    ("Receita", 4200),
    ("Despesa", 950),
    ("Receita", 3100),
    ("Despesa", 700)
]

Percorra a lista.

Mostre cada movimentação.

No final informe:

Quantidade de receitas.
Quantidade de despesas.
Valor total das receitas.
Valor total das despesas.
Se o total de receitas for maior que o total de despesas:
Empresa teve lucro.

Caso contrário:

Empresa teve prejuízo.
