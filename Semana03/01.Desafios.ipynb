```python
# Exercício 01 Aula
# Crie uma lista com 5 itens e imprima o segundo e o último usando índices.

print("=" * 40)
print("RELATÓRIO PRODUTOS")
print("=" * 40)

produtos = ["Feijão","Arroz","Carne","Cebola","Tomate"]

print(f"Quantidade de produtos: {len(produtos)}")
print(f"O primeiro produto da lista é: {produtos[0]}")
print(f"O último produto da lista é: {produtos[-1]}")
print("\nLista completa:")

for produto in produtos:
    print(f" - {produto}")
```

```python
# Exercício 1 - Vendas por Produto
# Uma loja registrou as seguintes vendas:
# Percorra as listas e mostre produto e quantidade
# Ao final informe: Quantos produtos venderam mais de 10 unidades. 
produtos = ["Notebook", "Mouse", "Teclado", "Monitor"] 
quantidades = [8, 25, 12, 5]
contador = 0

print("=" * 40)
print("RELATÓRIO VENDAS")
print("=" * 40)

for produto,quantidade in zip(produtos,quantidades):
    print(f"Produto ..... ",produto)
    print(F"Quantidade .... ",quantidade,"\n")


    if quantidade > 10:
        contador += 1

print(f"Produtos que venderam mais de 10 unidades: {contador}")
```

```python
# Exercício 2 - Cadastro de Funcionário
# Crie um dicionário para armazenar:
    # Nome
    # Cargo
    # Salário
    # Tempo de empresa

# Mostre todas as informações utilizando f-strings.
# Depois faça:
    # Se salário for maior que 5000 = Salário acima da média.
    # Caso contrário = Salário dentro da média.
    
funcionarios = {
    1:{"Nome": "João","Cargo": "Analista","Salário": 4500,"Tempo de empresa": 3},
    2:{"Nome": "Maria","Cargo": "Gerente","Salário": 7000,"Tempo de empresa": 5},
    3:{"Nome": "Pedro","Cargo": "Desenvolvedor","Salário": 5500,"Tempo de empresa": 2},
    4:{"Nome": "Ana","Cargo": "Designer","Salário": 4000,"Tempo de empresa": 1}
}

print("=" * 40)
print("CADASTRO FUNCIONÁRIOS")
print("=" * 40)

for codigo, funcionario in funcionarios.items():
    if funcionario['Salário'] > 5000:
        margem_salario = "Salário Acima da Média"
    
    else:
        margem_salario = "Salário Dentro da Média"
    
    print(f"\nFuncionário ..... {codigo}")
    print(f"Nome ..... {funcionario['Nome']}")
    print(f"Cargo ..... {funcionario['Cargo']}")
    print(f"Salário ..... R$ {funcionario['Salário']:.2f}")
    print(f"Tempo de empresa ..... {funcionario['Tempo de empresa']} anos")
    print(f"Margem Salário ..... {margem_salario}")
    print("=" * 5)
```
