```python
# Exercício 2 - Cadastro de Funcionário
# Crie um dicionário para armazenar:
    # Nome
    # Cargo
    # Salário
    # Tempo de empresa

# Mostre todas as informações utilizando f-strings.
# Depois faça:
    # Se salário for maior que 5000 = Salário acima da média.
    # Caso contrário = Salário dentro da média.
    
funcionarios = {
    1:{"Nome": "João","Cargo": "Analista","Salário": 4500,"Tempo de empresa": 3},
    2:{"Nome": "Maria","Cargo": "Gerente","Salário": 7000,"Tempo de empresa": 5},
    3:{"Nome": "Pedro","Cargo": "Desenvolvedor","Salário": 5500,"Tempo de empresa": 2},
    4:{"Nome": "Ana","Cargo": "Designer","Salário": 4000,"Tempo de empresa": 1}
}

print("=" * 40)
print("CADASTRO FUNCIONÁRIOS")
print("=" * 40)

for codigo, funcionario in funcionarios.items():
    if funcionario['Salário'] > 5000:
        classificacao_salario = "Salário Acima da Média"
    
    else:
        classificacao_salario = "Salário Dentro da Média"
    
    print(  f"\nFuncionário ..... {codigo}"
            f"\nNome ..... {funcionario['Nome']}"
            f"\nCargo ..... {funcionario['Cargo']}"
            f"\nSalário ..... R$ {funcionario['Salário']:.2f}"
            f"\nTempo de empresa ..... {funcionario['Tempo de empresa']} ano(s)"
            f"\nClassificação Salário ..... {classificacao_salario}"
            )
    print("=" * 40)
```

```python
# Exercício 3 - Controle de Estoque
# Uma empresa possui estoque
estoque = {
    "Notebook": 18,
    "Mouse": 4,
    "Monitor": 7,
    "Teclado": 25
}

# Percorra e para cada produto:
    # Se quantidade menor que 5 = Reposição urgente
    # Caso contrário = Estoque OK 

print("=" * 40)
print("CONTROLE REPOSIÇÃO ESTOQUE")
print("=" * 40)

reposicao = 0
estoque_ok = 0

for produto, quantidade in estoque.items():

    if quantidade < 5:
        status = "Reposição urgente"
        reposicao += 1
    elif quantidade < 10:
        status = "Estoque baixo"
        estoque_ok +=1
    else:
        status = "Estoque OK"
        estoque_ok +=1


    print(f"{produto}: {status} ({quantidade} itens)")

print(
    f"=" * 40,
    f"\n\nProdutos com reposição urgente: {reposicao}"
    f"\nProdutos com estoque OK: {estoque_ok}")
```

```python
# Exercício 4 - Análise Comercial
# A empresa possui uma base de vendas.
# Percorra a base e informe:
    # Quantas vendas foram maiores que R1000 
    # Quantas vendas são valores pares (utilize %)
# Ao final mostre o total de cada categoria.

vendas = [1250, 980, 2300, 450, 1750, 890, 3201]

print("=" * 40)
print("RELATÓRIO VENDAS")
print("=" * 40)

venda_alta = 0
venda_baixa = 0
venda_par = 0
total_vendas_alta = 0
total_vendas_baixa = 0
faturamento_total = sum(vendas)
quantidade_vendas = len(vendas)
maior_venda = max(vendas)
menor_venda = min(vendas)

for venda in vendas:
    if venda > 1000:
        venda_alta += 1
        total_vendas_alta += venda
    else:
        venda_baixa += 1
        total_vendas_baixa += venda
    
    if venda %2 == 0:
        venda_par += 1

percentual_venda_alta = (total_vendas_alta / faturamento_total) * 100
percentual_venda_baixa = (total_vendas_baixa / faturamento_total) * 100

print(
    f"VENDAS ACIMA DE R$ 1.000"
    f"\nQuantidade ..... {venda_alta}"
    f"\nTotal ..... R$ {total_vendas_alta:.2f}"
    f"\nPercentual ..... {percentual_venda_alta:.2f}%"

    f"\n\nVENDAS ABAIXO DE R$ 1.000"
    f"\nQuantidade ..... {venda_baixa}"
    f"\nTotal ..... R$ {total_vendas_baixa:.2f}"
    f"\nPercentual ..... {percentual_venda_baixa:.2f}%"

    f"\n\nRESULTADO GERAL"
    f"\nTotal vendas ..... R$ {faturamento_total:.2f}"
    f"\nMaior vendas ..... R$ {maior_venda:.2f}"
    f"\nMenor vendas ..... R$ {menor_venda:.2f}"
    f"\nVendas Par ..... {venda_par}"
    f"\nTotal de Vendas ..... {quantidade_vendas}"
    )
```

```python
# Exercício 5 - Ranking de Lojas
# Uma empresa possui uma base.
# Percorra a lista.
# Para cada loja:
    # Se faturamento maior que R$100.000 = Meta atingida
    # Caso contrário = Meta não atingida

lojas = [
         {"nome": "Joinville", "faturamento": 85000},
         {"nome": "Curitiba", "faturamento": 112000},
         {"nome": "Blumenau", "faturamento": 76000},
         {"nome": "Florianópolis", "faturamento": 135000}
         ]
meta = 100000
qtd_acima_meta = 0
qtd_abaixo_meta = 0
faturamento_total = 0

print("=" * 40)
print("RELATÓRIO VENDAS POR LOJA")
print("=" * 40)

for loja in lojas:
  if loja["faturamento"] >= meta:
    atingimento_meta = "Sim"
    qtd_acima_meta += 1
  else:
    atingimento_meta = "Não"
    qtd_abaixo_meta +=1

  print(f"\nLoja: {loja['nome']}")
  print(f"Faturamento: R$ {loja['faturamento']:.2f}")
  print(f"Atingiu a Meta? {atingimento_meta}")

for loja in lojas:
  faturamento_total += loja["faturamento"]

print("=" * 40)
print("RESUMO")
print(f"Faturamento Total: R$ {faturamento_total:.2f}")
print(f"Quantidade Lojas Acima da Meta: {qtd_acima_meta}")
print(f"Quantidade Lojas Abaixo da Meta: {qtd_abaixo_meta}")
```

```python
# Exercício 6 - Auditoria de Clientes
# Cada tupla representa: (nome, quantidade de compras)
# Mostre:
    # Felipe realizou 5 compras.
# Depois classifique:
    #  0 compras = Cliente Inativo
    # 1 até 3 = Cliente Ativo
    # Acima de 3 = Cliente VIP

clientes = [
    ("Felipe", 5),
    ("Ana", 1),
    ("João", 0),
    ("Carlos", 3),
    ("Maria", 8) ]

print("=" * 40)
print("AUDITORIA DE CLIENTES")
print("=" * 40)

for cliente in clientes:
  if cliente[1] > 3:
    classificacao = "Cliente VIP"
  elif cliente[1] >= 1:
    classificacao = "Cliente Ativo"
  else:
    classificacao = "Cliente Inativo"

  print(f"- {cliente[0]} realizou {cliente[1]} compras. {classificacao}.")
```

```python
# Exercício 7 - Dashboard Comercial
# Percorra a lista e informe:
# Quantas vendas cada vendedor realizou.
# Quantas vendas foram maiores que R1500.
vendas = [
    {"vendedor":"Ana","valor":1200},
    {"vendedor":"Ana","valor":800},
    {"vendedor":"Carlos","valor":2100},
    {"vendedor":"Maria","valor":950},
    {"vendedor":"Carlos","valor":1800},
    {"vendedor":"Maria","valor":1500}
]

contador_vendas = {}
vendas_altas = 0

for venda in vendas:
    vendedor = venda["vendedor"]

    # Conta quantas vendas cada vendedor fez
    if vendedor in contador_vendas:
        contador_vendas[vendedor] += 1
    else:
        contador_vendas[vendedor] = 1

    # Conta vendas acima de 1500
    if venda["valor"] > 1500:
        vendas_altas += 1

print("=" * 40)
print("DASHBOARD COMERCIAL")
print("=" * 40)

for vendedor, quantidade in contador_vendas.items():
    print(f"{vendedor} realizou {quantidade} vendas.")

print(f"\nVendas acima de R$ 1500: {vendas_altas}")
```

```python
# Exercício 8 - Controle de Chamados de TI
# Para cada situação:
    # Se quantidade maior que 10 = Alta demanda
    # Caso contrário = Fluxo normal

# Depois informe quantos chamados ainda não foram resolvidos.
# Considere "Aberto" e "Em andamento" como não resolvidos.
# Uma empresa possui:

chamados = { "Aberto": 14, "Em andamento": 7, "Resolvido": 31, "Cancelado": 2 }
encerrado = 0
aberto = 0

print("=" * 40)
print("DASHBOARD CHAMADOS TI")
print("=" * 40)

for chamado,quantidade in chamados.items():
  if quantidade > 10:
    demanda = "Alta Demanda"
  else:
    demanda = "Fluxo Normal"
  print(f"{chamado} | Quantidade: {quantidade} | {demanda}")
  if chamado in ["Aberto", "Em andamento"]:
    aberto += quantidade
  else:
    encerrado += quantidade
print("\nRESUMO")
print(f"Quantidade 'Em Aberto': {aberto}")
print(f"Quantidade 'Encerrado': {encerrado}")
```

```python
# Exercício 9 - Controle Financeiro
# Uma empresa registrou as movimentações:
# Percorra a lista. Mostre cada movimentação.
# No final informe:
    # Quantidade de receitas.
    # Quantidade de despesas.
    # Valor total das receitas.
    # Valor total das despesas.
# Se o total de receitas for maior que o total de despesas: Empresa teve lucro.
# Caso contrário: Empresa teve prejuízo.

movimentacoes = [
                 ("Receita", 2500),
                 ("Despesa", 1200),
                 ("Receita", 4200),
                 ("Despesa", 950),
                 ("Receita", 3100),
                 ("Despesa", 700)
                 ]
receita = 0
qtd_receita = 0
despesa = 0
qtd_despesa = 0

print("=" * 40)
print("CONTROLE FINANCEIRO")
print("=" * 40)

for movimentacao in movimentacoes:
  valor = movimentacao[1]
  tipo = movimentacao[0]

  if tipo == "Receita":
    receita += valor
    qtd_receita += 1
  else:
    despesa += valor
    valor = -valor
    qtd_despesa += 1

  print(f"{tipo} | R$ {valor:.2f}")

if receita > despesa:
    resultado_final = "LUCRO"
else:
    resultado_final ="PREJUÍZO"

saldo = receita - despesa

print()
print("=" * 40)
print("RESUMO")
print(f"Receita | Quantidade: {qtd_receita} | Valor: R$ {receita:.2f}")
print(f"Despesa | Quantidade: {qtd_despesa} | Valor: R$ {despesa:.2f}")
print(f"\nSaldo Final ..... R$ {saldo:.2f}")
print(f"Ao final do período a empresa teve: {resultado_final }")
```

```python
# Exercício 9
# Faça um relatórico completo
# estoque < 10 = Reposição Urgente
# estoque >= 10 = Estoque OK

estoque = {
    "Notebook":{"Preço":3500,"Quantidade":8},
    "Mouse":{"Preço":80,"Quantidade":25},
    "Monitor":{"Preço":1200,"Quantidade":3},
    "Teclado":{"Preço":150,"Quantidade":12}
}

qtd_produtos = 0
qtd_produtos_urgente = 0
valor_estoque = 0


print("=" * 40)
print("RELATÓRIO DE ESTOQUE")
print("=" * 40)


for produto, dados in estoque.items():
    valor = dados['Preço']
    quantidade = dados['Quantidade']
    qtd_produtos += quantidade
    valor_estoque += valor * quantidade
    
    if quantidade >= 10:
        status_estoque = "OK"
    else:
        status_estoque = "Reposição URGENTE!"
        qtd_produtos_urgente += 1
        
    print(f"\nProduto: {produto}")
    print(f"Preço: R$ {valor:.2f}")
    print(f"Quantidade: {quantidade}")
    print(f"Estoque: {status_estoque}")

print()
print("=" * 40)
print(f"Total de produtos cadastrados: {len(estoque)}")
print(f"Produtos com reposição urgente: {qtd_produtos_urgente}")
print(f"Quantidade produtos em estoque: {qtd_produtos}")
print(f"Valor total do estoque: R$ {valor_estoque:.2f}")
```

```python
vendedores = [
    {"Nome":"Ana","Venda":1800},
    {"Nome":"Carlos","Venda":900},
    {"Nome":"Maria","Venda":2400},
    {"Nome":"João","Venda":700}
]

venda_total = 0

maior_venda = vendedores[0]["Venda"]
menor_venda = vendedores[0]["Venda"]

print("=" * 40)
print("RELATÓRIO VENDAS")
print("=" * 40)

for vendedor in vendedores:

    nome = vendedor["Nome"]
    venda = vendedor["Venda"]

    venda_total += venda

    if venda > maior_venda:
        maior_venda = venda

    if venda < menor_venda:
        menor_venda = venda

    print(f"{nome} .......... R$ {venda:.2f}")

media = venda_total / len(vendedores)

qtd_maior_media = 0
qtd_menor_media = 0

for vendedor in vendedores:

    if vendedor["Venda"] > media:
        qtd_maior_media += 1

    elif vendedor["Venda"] < media:
        qtd_menor_media += 1

print()
print("=" * 40)
print(f"Total de Vendas ..... R$ {venda_total:.2f}")
print(f"Maior Venda ......... R$ {maior_venda:.2f}")
print(f"Menor Venda ......... R$ {menor_venda:.2f}")
print(f"Média das Vendas .... R$ {media:.2f}")
print(f"Acima da Média ...... {qtd_maior_media}")
print(f"Abaixo da Média ..... {qtd_menor_media}")
```
