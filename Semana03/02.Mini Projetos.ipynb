Exercício 10 — Mini Projeto de Análise de Dados (Desafiador)

Você recebeu uma base simplificada de vendas:

vendas = [
    {"produto":"Notebook","categoria":"Informática","valor":4200},
    {"produto":"Mouse","categoria":"Informática","valor":80},
    {"produto":"Mesa","categoria":"Móveis","valor":950},
    {"produto":"Monitor","categoria":"Informática","valor":1400},
    {"produto":"Cadeira","categoria":"Móveis","valor":780},
    {"produto":"Notebook","categoria":"Informática","valor":3900}
]

Sem utilizar funções prontas como sum() ou max(), percorra a lista e descubra:

Quantas vendas existem.
Quantos produtos pertencem à categoria Informática.
Quantos pertencem à categoria Móveis.
Quantas vendas foram acima de R$1.000.
Quantas vendas foram abaixo ou iguais a R$1.000.
Valor total vendido (utilizando um acumulador).
Informe se a empresa bateu a meta:
Meta atingida: faturamento acima de R$10.000.
Meta não atingida: caso contrário.
