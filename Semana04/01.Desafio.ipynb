```python
# Leitura arquivo CSV
import csv

with open("cadastro_funcionario.csv", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    funcionarios = list(leitor)
print(f"Total de funcionários: {len(funcionarios)}") # quantidade de linhas
print(funcionarios[0]) # mostrar a primeira linha


# Leitura arquivo JSON
import json

with open("emails_corporativos.json", "r", encoding="utf-8") as arquivo:
    emails = json.load(arquivo)
print(f"\nTotal de funcionários: {len(emails)}") # quantidade de linhas
print(emails[0]) # mostrar a primeira linha
```
