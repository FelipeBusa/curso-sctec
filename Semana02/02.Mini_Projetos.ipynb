# Mini Projeto 2
Uma empresa possui estas vendas: 
'vendas = [1200, 850, 930, 2000, 1570]'

Faça um programa que mostre:
- total vendido; 
- maior venda; 
- menor venda; 
- média das vendas;
- Diferença entre maior e menor
- Quantidade de vendas acima da média
- Quantidade de vendas abaixo da média

```python
vendas = [1200, 850, 930, 2000, 1570] 

print("MODELO 1 - UTILIZANDO FORMULAS PRONTAS")
print("=" * 40)
print("RELATÓRIO DE VENDAS")
print("=" * 40)

total_vendas = sum(vendas)
total_produtos = len(vendas)
maior_venda = max(vendas)
menor_venda = min(vendas)
media_vendas = total_vendas / total_produtos
dif_vendas = maior_venda - menor_venda
qtd_maior_media = 0
qtd_menor_media = 0

for venda in vendas:
  if venda > media_vendas:
    qtd_maior_media += 1

  if venda < media_vendas:
    qtd_menor_media += 1


print()
print(f"Total de Vendas","."*5,f"R$ {total_vendas:.2f}")
print(f"Quantidade de Vendas","."*5,f"{total_produtos}")
print(f"Média das Vendas","." * 5,f"R$ {media_vendas:.2f}")
print(f"Maior Venda","." * 5,f"R$ {maior_venda:.2f}")
print(f"Menor Venda","." * 5,f"R$ {menor_venda:.2f}")
print(f"Diferença Entre Maior e Menor Vendas","." * 5,f"R$ {dif_vendas:.2f}")
print(f"Qtd Vendas Maior que a Média","." * 5,f"{qtd_maior_media}")
print(f"Qtd Vendas Menor que a Média","." * 5,f"{qtd_menor_media}")



# Modelo 2 - sem utilizar as formulas sum(), max() e min()
print("\n\nMODELO 2 - NÃO UTILIZANDO FORMULAS PRONTAS")
print("=" * 40)
print("RELATÓRIO DE VENDAS")
print("=" * 40)

venda_total = 0
venda_maior = vendas[0]
venda_menor = vendas[0]
maior_media_qtd = 0
menor_media_qtd = 0

for venda in vendas:
  if venda < venda_menor:
    venda_menor = venda

  elif venda > venda_maior:
    venda_maior = venda

  venda_total += venda


venda_media = venda_total / len(vendas)
venda_dif = venda_maior - venda_menor

for venda in vendas:
  if venda > venda_media:
    maior_media_qtd += 1

  elif venda < venda_media:
    menor_media_qtd += 1

print()
print(f"Total de Vendas","."*5,f"R$ {venda_total:.2f}")
print(f"Quantidade de Vendas","."*5,f"{len(vendas)}")
print(f"Média das Vendas","." * 5,f"R$ {venda_media:.2f}")
print(f"Maior Venda","." * 5,f"R$ {venda_maior:.2f}")
print(f"Menor Venda","." * 5,f"R$ {venda_menor:.2f}")
print(f"Diferença Entre Maior e Menor Vendas","." * 5,f"R$ {venda_dif:.2f}")
print(f"Qtd Vendas Maior que a Média","." * 5,f"{maior_media_qtd}")
print(f"Qtd Vendas Menor que a Média","." * 5,f"{menor_media_qtd}")
```

# Mini Projeto 3
Considerando o Mini Projeto 2. Classifique cada venda como:
- Alta → acima da média
- Baixa → abaixo da média
- Na Média → igual à média (se existir)

```python
print("=" * 40)
print("RELATÓRIO DE VENDAS")
print("=" * 40)

vendas = [1200, 850, 930, 2000, 1570] 
venda_media = sum(vendas) / len(vendas)
venda_ordem = 0
maior_media_qtd = 0
menor_media_qtd = 0
na_media_qtd = 0

for venda in vendas:
  if venda > venda_media:
    classificacao = "Alta"
    maior_media_qtd += 1
    
  elif venda < venda_media:
    classificacao = "Baixa"
    menor_media_qtd += 1
      
  else:
    classificacao = "Média"
    na_media_qtd += 1

  venda_ordem += 1

  print(f"Venda {venda_ordem} ..... R$ {venda:.2f}")
  print(f"Classificação ..... {classificacao}\n")
print("=" * 40)

print(f"Média das vendas ..... R$ {venda_media}")
print(f"Qtd. Vendas Maior que a Média ..... {maior_media_qtd}")
print(f"Qtd. Vendas Menor que a Média ..... {menor_media_qtd}")
print(f"Qtd. Vendas na Média ..... {na_media_qtd}")
```

# Mini Projeto 4
Replique o Mini Projeto 3, considerando o dicionário abaixo e mostrando resultado por vendedor:

```python
vendedores = {
    "Ana": 1200,
    "Bruno": 850,
    "Carlos": 930,
    "Diego": 2000,
    "Elisa": 1570
}

total_vendas = sum(vendedores.values())
media_vendas = total_vendas / len(vendedores)
maior_media_qtd = 0
menor_media_qtd = 0
na_media_qtd = 0

print("=" * 40)
print("RELATÓRIO DE DESEMPENHO")
print("=" * 40)

for nome, vendas in vendedores.items():

  if vendas > media_vendas:
    maior_media_qtd += 1
    classificacao = "Alta"
  
  elif vendas < media_vendas:
    menor_media_qtd += 1
    classificacao = "Baixa"
      
  else:
    classificacao = "Média"
    na_media_qtd += 1

  print(f"\nVendedor(a) ..... {nome}")
  print(f"Vendas ..... R$ {vendas:.2f}")
  print(f"Classificação ..... {classificacao}")

print("\n" + "=" * 40)
print(f"Média das vendas ..... R$ {media_vendas:.2f}")
print(f"Qtd. Vendas Maior que a Média ..... {maior_media_qtd}")
print(f"Qtd. Vendas Menor que a Média ..... {menor_media_qtd}")
print(f"Qtd. Vendas na Média ..... {na_media_qtd}")
```
