## Exercícios da Semana
* Fácil: Crie 3 variáveis (uma int, uma float e uma str) com dados reais seus e confirme o tipo de cada uma com type().
* Médio: Escreva um programa com if/elif/else que classifique uma idade em "Criança", "Adolescente" ou "Adulto".
* Avançado: Escreva um programa que percorra os números de 1 a 20 e identifique quais são múltiplos de 3, de 5, de ambos ou de nenhum dos dois.
