```python
# Exercício 1
# Crie uma lista com 10 preços de produtos e calcule o valor total da compra.

preco = [10,5,9,45,87,64,15,37,99,1]

valor_total = 0
for i in preco:
    valor_total +=i
print(f"O valor total da compra é: R$ {valor_total:.2f}")
```
