```python
# Exercício 1
# Crie uma lista com 10 preços de produtos e calcule o valor total da compra.
precos_produtos = [10,5,9,45,87,64,15,37,99,1]

# Modelo 1
print("=" * 40)
print("Modelo 1 - Valor Total Compra")
valor_total = 0
for valor in precos_produtos:
    valor_total += valor
print(f"O valor total da compra é: R$ {valor_total:.2f}")

# Modelo 2 - Estilo de Lista de Produtos
print("\n","=" * 40)
print("Modelo 2 - Lista de Produtos") # apenas para separar os resultados

precos_produtos = [10,5,9,45,87,64,15,37,99,1]
valor_total = 0
nome_produto = 0 # Criado para dar nome aos produtos na ordem

print("A lista de produtos com os seus respectivos preços:")
for valor_produto in precos_produtos:
    nome_produto += 1 # criado para alterar o nome do produto na ordem
    print(f" - Produto {nome_produto} tem o valor: R$ {valor_produto:.2f}")

for valor in precos_produtos:
    valor_total += valor
print(f"\nO valor total da compra é: R$ {valor_total:.2f}")



# O que aprendi neste exercício? 
# Podem existir várias formas de realizar o que foi preposto.
# O interessante é sempre procurar o que for mais adequado e 'simples' para fazer.

# Qual solução achei mais eficiente?
# O modelo 2 é o mais adequado para apresentar uma lista com todos os produtos e valores.
# Ainda dá para ajustar e passar apenas 1 vez no for e trazer o valor individual e o valor final
# O modelo 1 é o mais adequando para um resultado final, mais simples e rápido.

# O que faria diferente em uma próxima versão?
# Ajustar o modelo 2 para 1 for apenas e resolver todas as necessidades.
```

```python
# Exercício 3
# Dada a lista: notas = [8.5, 7.0, 9.5, 6.0, 10]. Calcule a média.

notas = [8.5, 7.0, 9.5, 6.0, 10]

# Modelo 1 - Utilizando a formula sum direto
print("=" * 40)
print("Modelo 1 - Utilizando a formula sum() direto")
print(f"Foram contabilizadas {len(notas)} notas e o total das notas é {sum(notas):.2f}.")
print(f"A média das notas é {(sum(notas) / len(notas)):.2f}.\n")

# Modelo 2 - Utilizando for() para criar o cálculo
print("=" * 40)
print("Modelo 2 - Utilizando for() para criar o cálculo")

total = 0
quantidade = len(notas)

for nota in notas:
    total += nota
print(f"Foram contabilizadas {quantidade} notas e o total das notas é {total:.2f}.")

media = total / quantidade
print(f"A média das notas é {media:.2f}.\n")


# Modelo 3 - Utilizando for() para criar o cálculo
print("=" * 40)
print("Modelo 3 - RELATÓRIO DE NOTAS")

total = sum(notas)
quantidade = len(notas)
media = total / quantidade

print(f"Quantidade de notas: {quantidade}")
print(f"Soma das notas: {total:.2f}")
print(f"Média da turma: {media:.2f}")



# O que aprendi neste exercício?
# A formula sum já simplifica muito o cálculo 
# E dentro do print podemos colocar o resultado da váriavel sem precisar criar uma.

# Qual solução achei mais eficiente? 
# O modelo 3 é mais eficiente, simples e clean.

# O que faria diferente em uma próxima versão? 
# No momento não encontro o que poderá ser feito de diferente. 
```

```python
# Exercício 4
# Dada a lista: idades = [15, 42, 19, 65, 12, 30, 18] Conte quantas pessoas são maiores de idade.

idades = [15, 42, 19, 65, 12, 30, 18]

quantidade_pessoas = len(idades)
maiores_idade = 0

for idade in idades:
    if idade >= 18:
        maiores_idade += 1
        
menores_idade = quantidade_pessoas - maiores_idade
media_idade = sum(idades) / quantidade_pessoas
percentual_maiores = (maiores_idade / quantidade_pessoas) * 100
percentual_menores = 100 - percentual_maiores

print("=" * 40)
print("RELATÓRIO DE IDADES")
print(f"Total de pessoas: {quantidade_pessoas}")
print(f"Maiores de idade: {maiores_idade} (18 anos ou mais). Representam: {percentual_maiores:.0f}% do total")
print(f"Menores de idade: {menores_idade} (até 17 anos).Representam: {percentual_menores:.0f}% do total")
print(f"A média de idade do grupo: {media_idade:.0f} anos.")


# O que aprendi neste exercício? 
# Interpretar melhor o pedido e trazer mais informações que podem ser uteis.

# Qual solução achei mais eficiente?
# Apenas a solução apresentada.

# O que faria diferente em uma próxima versão?
# Entendo que não tenha o que ajustar.
```

```python
# Exercício 5
# Encontre o menor valor da lista abaixo sem usar min(). temperaturas = [28, 34, 18, 25, 40, 12]
temperaturas = [28, 34, 18, 25, 40, 12]

print("=" * 40)
print("Modelo 1 - NÃO utilizando min()")

menor_temperatura = temperaturas[0]
maior_temperatura = temperaturas[0]

for temperatura in temperaturas:
    if temperatura < menor_temperatura:
        menor_temperatura = temperatura
    if temperatura > maior_temperatura:
        maior_temperatura = temperatura

variacao_temperatura = maior_temperatura - menor_temperatura
        
print(f"A menor temperatura registrada é de {menor_temperatura:.0f}º")
print(f"A maior temperatura registrada é de {maior_temperatura:.0f}º")
print(f"A variação de temperatura foi de {variacao_temperatura:.0f}º\n")

 
print("=" * 40)
print("Modelo 2 - Utilizando min()")

temperatura_minima = min(temperaturas)
temperatura_maxima = max(temperaturas)
temperatura_variacao = temperatura_maxima - temperatura_minima

print(f"A menor temperatura registrada é de {temperatura_minima:.0f}º")
print(f"A maior temperatura registrada é de {temperatura_maxima:.0f}º")
print(f"A variação de temperatura foi de {temperatura_variacao:.0f}º")



# O que aprendi neste exercício?
# A calcular o menor e maior sem precisar das formulas min() e max()

# Qual solução achei mais eficiente?
# Utilizando a formula min() pronta.

# O que faria diferente em uma próxima versão?
# Entendo que não tenha o que ajustar.
```

```python
# Exercício 9
# Peça números ao usuário. Enquanto ele não digitar 0, continue somando. No final mostre a soma.

# Modelo 1 - Soma de Números
print("=" * 40)
print("MODELO 2 - SOMA DE NÚMEROS")

soma = 0

numero = int(input("Digite um número (0 para sair): "))

while numero != 0:
    soma += numero
    numero = int(input("Digite outro número (0 para sair): "))

print(f"\nSoma total: {soma}")


# Modelo 2 - Criando um jogo a partir desse exercício
print("=" * 40)
print("MODELO 2 - JOGO DESCUBRA O NÚMERO SECRETO")

numero = int(input("Descubra o número secreto! Digite um número ENTRE 0 e 9: "))
soma = 0
tentativa = 0

while numero != 0:
  soma += numero
  tentativa += 1
  print(f"\nNÃO ACERTOU! \nFavor digitar novamente um número ENTRE 0 e 9.")
  print(f"Número informado: {numero}")
  print(f"Número de tentativas: {tentativa}")
  print(f"Soma total das tentativas: {soma}")
  numero = (int(input("Digite seu número: ")))

print("\nVOCÊ ACERTOU! Parabéns!")
print(f"Total de tentativas: {tentativa+1}")

# O que aprendi neste exercício?
# A composição de While com vários cenários (if, elif e break)

# Como utilizar o input e também transformar
# Nesse exercício além do input foi necessário transformar o dado para número inteiro int()

# O que faria diferente em uma próxima versão?
# Acredito que ainda dé para melhorar o While
```

```python
# Exercício 10 # Sistema de Caixa. Uma loja possui esta lista de compras: 
# compras = [25, 40, 18, 90, 12, 35] 
# Use um for para calcular o valor total das compras. 
# Depois use um while para simular pagamentos de R 50emR 50 até quitar toda a dívida. 
# Mostre quanto falta pagar após cada pagamento.

compras = [25, 40, 18, 90, 12, 35]
pagamento = 50
quantidade_pagamento = 0
total_compras = 0
numero_produto = 0 # Criado para dar número aos produtos na ordem


print("=" * 40)
print("SISTEMA DE CAIXA")
print("=" * 40)
for valor_produto in compras:
    numero_produto += 1 # criado para alterar o nome do produto na ordem
    print(f" - Produto {numero_produto}","." * 10,f"R$ {valor_produto:.2f}")
print("=" * 40)


for compra in compras:
    total_compras += compra
ticket_medio = total_compras/len(compras)
print("\nTotal compras","." * 8,f"R$ {total_compras:.2f}")


while total_compras >= pagamento:
    quantidade_pagamento += 1
    total_compras -= pagamento
    print(f"\nPagamento {quantidade_pagamento}","." * 11,f"R$ {pagamento:.2f}")
    print("Saldo","." * 16,f"R$ {total_compras:.2f}")


    if total_compras < pagamento:
        quantidade_pagamento +=1
        ultimo_pagamento = total_compras
        total_compras -= ultimo_pagamento
        print("\nPagamento Final","." * 5,f"R$ {ultimo_pagamento:.2f}\n")


print(f"=" * 40)

print(f"PAGAMENTOS realizados","." * 10,f"{quantidade_pagamento}")
print(f"\nTicket Médio","." * 15,f"{ticket_medio:.2f}")
print(f"\nProdutos comprados","." * 13,f"{len(compras)}")


# O que aprendi neste exercício?
# Como um sistema simples é feito e como pode melhorar cada vez mais

# O que faria diferente em uma próxima versão?
# Acredito que dê para ajustar melhor o código, deixar mais clean
```
