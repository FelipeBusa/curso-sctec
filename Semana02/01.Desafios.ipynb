```python
# Exercício 1
# Crie uma lista com 10 preços de produtos e calcule o valor total da compra.

precos_produtos = [10,5,9,45,87,64,15,37,99,1]
print("=" * 40)
print("Modelo 1") # Apenas para separar os modelos
valor_total = 0
for valor in precos_produtos:
    valor_total += valor
print(f"O valor total da compra é: R$ {valor_total:.2f}")

# Agora mostrando a lista de produtos com cada preço e o valor total final:
print("=" * 40)
print("Modelo 2") # apenas para separar os resultados

precos_produtos = [10,5,9,45,87,64,15,37,99,1]
valor_total = 0
nome_produto = 0 # Criado para dar nome aos produtos na ordem

print("A lista de produtos com os seus respectivos preços:")
for valor_produto in precos_produtos:
    nome_produto += 1 # criado para alterar o nome do produto na ordem
    print(f" - Produto {nome_produto} tem o valor: R$ {valor_produto:.2f}")

for valor in precos_produtos:
    valor_total += valor
print(f"O valor total da compra é: R$ {valor_total:.2f}")

# O que aprendi neste exercício? 
# Podem existir várias formas de realizar o que foi preposto.
# O interessante é sempre procurar o que for mais adequado e 'simples' para fazer.

# Qual solução achei mais eficiente?
# O modelo 2 é o mais adequado para apresentar uma lista com todos os produtos e valores.
# Ainda dá para ajustar e passar apenas 1 vez no for e trazer o valor individual e o valor final
# O modelo 1 é o mais adequando para um resultado final, mais simples e rápido.

# O que faria diferente em uma próxima versão?
# Ajustar o modelo 2 para 1 for apenas e resolver todas as necessidades.
```

```python
# Exercício 3
# Dada a lista: notas = [8.5, 7.0, 9.5, 6.0, 10]. Calcule a média.

notas = [8.5, 7.0, 9.5, 6.0, 10]

# Modelo 1 - Utilizando a formula sum direto
print("=" * 40)
print("Modelo 1")
print(f"Foram contabilizadas {len(notas)} notas e o total das notas é {sum(notas):.2f}.")
print(f"A média das notas é {(sum(notas) / len(notas)):.2f}.")

# Modelo 2 - Utilizando for() para criar o cálculo
print("=" * 40)
print("Modelo 2")

total = 0
quantidade = len(notas)

for nota in notas:
    total += nota
print(f"Foram contabilizadas {quantidade} notas e o total das notas é {total:.2f}.")

media = total / quantidade
print(f"A média das notas é {media:.2f}.")


# Modelo 3 - Utilizando for() para criar o cálculo
print("=" * 40)
print("Modelo 3 - RELATÓRIO DE NOTAS")

total = sum(notas)
quantidade = len(notas)
media = total / quantidade

print(f"Quantidade de notas: {quantidade}")
print(f"Soma das notas: {total:.2f}")
print(f"Média da turma: {media:.2f}")

# O que aprendi neste exercício?
# A formula sum já simplifica muito o cálculo 
# E dentro do print podemos colocar o resultado da váriavel sem precisar criar uma.

# Qual solução achei mais eficiente? 
# O modelo 3 é mais eficiente, simples e clean.

# O que faria diferente em uma próxima versão? 
# No momento não encontro o que poderá ser feito de diferente. 
```

```python
# Exercício 4
# Dada a lista: idades = [15, 42, 19, 65, 12, 30, 18] Conte quantas pessoas são maiores de idade.

idades = [15, 42, 19, 65, 12, 30, 18]

quantidade_pessoas = len(idades)
maiores_idade = 0
for idade in idades:
    if idade >= 18:
        maiores_idade += 1
        
menores_idade = quantidade_pessoas - maiores_idade
media_idade = sum(idades) / quantidade_pessoas
percentual_maiores = (maiores_idade / quantidade_pessoas) * 100
percentual_menores = 100 - percentual_maiores

print("=" * 40)
print("RELATÓRIO DE IDADES")
print("=" * 40)
print(f"Total de pessoas: {quantidade_pessoas}")
print(f"Maiores de idade: {maiores_idade} (18 anos ou mais). Representam: {percentual_maiores:.0f}% do total")
print(f"Menores de idade: {menores_idade} (até 17 anos).Representam: {percentual_menores:.0f}% do total")
print(f"A média de idade do grupo: {media_idade:.0f} anos.")


# O que aprendi neste exercício? 
# Interpretar melhor o pedido e trazer mais informações que podem ser uteis.

# Qual solução achei mais eficiente?
# Apenas a solução apresentada.

# O que faria diferente em uma próxima versão?
# Entendo que não tenha o que ajustar.
```

```python
# Exercício 7
# Comece com: saldo = 1000 Enquanto o saldo for maior que 0, retire R$150 e mostre o saldo restante.
saldo = 1000
saque = 150

print("=" * 40)
print("SALDO CONTA BANCÁRIA")
print(f"Saldo: R$ {saldo:.2f}")

while saldo >= saque:
    saldo -= saque
    print(f"====\nSaque de R$ {saque:.2f}.\nSaldo de R$ {saldo:.2f}")
print(f"\n====\nOPERAÇÃO ENCERRADA!\nSaldo insuficiente para saque.\nSaldo de R$ {saldo:.2f}")


# O que aprendi neste exercício? 
# Utilizar print() dentro e fora do while

# O que faria diferente em uma próxima versão?
# Criação de campos para input() do valores para depositar e valores a sacar. 
```
