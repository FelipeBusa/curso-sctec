## Mini Projeto 1

Você é analista comercial.

Crie um programa que:

receba o nome do vendedor;
receba o valor vendido;
se vender mais de R$ 50.000, informe que ganhou bônus;
caso contrário, informe que não ganhou.

## Mini Projeto 2

Uma empresa possui estas vendas: vendas = [1200, 850, 930, 2000, 1570]
Faça um programa que mostre:

total vendido;
maior venda;
menor venda;
média das vendas.
